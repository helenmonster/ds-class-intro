### Exercise 5. Find the mean and standard deviation of math and verbal SAT score for men-only, women-only, and non gender specific universities.

```python
import pandas as pd
df=pd.read_csv("data/college.csv")
df
```

```python
import numpy as np
df['gender']=np.where(df['menonly']==1, 'menonly', np.where(df['womenonly']==1, 'womenonly',np.where(df['menonly'].isna(),'missing',np.where(df['womenonly'].isna(), 'missing', 'notsure'))))
# df['gender'].value_counts()
df.groupby('gender').agg({'satvrmid': ['mean', 'std'],
                         'satmtmid': ['mean', 'std']})
```
